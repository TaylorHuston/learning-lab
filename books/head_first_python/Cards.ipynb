```python
import random

suits = ["Clubs", "Diamonds", "Hearts", "Spades"]
faces = ["Jack", "Queen", "King", "Ace"]
numbered = [2, 3, 4, 5, 6, 7, 8, 9, 10]
```

```python
suits
```

```python
def draw():
    the_suit = random.choice(suits)
    the_card = random.choice(faces + numbered)
    return f"{the_card} of {the_suit}"
```

```python
for _ in range(5):
    print(draw())
```

```python
deck = set()
print(type(deck))
print(len(deck))
print(dir(deck))
```

```python
help(deck.add)
```

```python
for suit in suits:
    for card in faces + numbered:
        deck.add((card, "of", suit))
print(len(deck))
```

```python
def draw():
    card = random.choice(list(deck))
    #print(type(card))
    #print(card)
    deck.remove(card)
    #print(len(deck))
    return card
```

```python
print(draw())
if card in deck:
    print("Card is still in deck")
else:
    print("Card has been removed from deck")
```
